#### Main Script

Runs a full simulation for scenario 2 (approximating to current easter bush energy system).

Then runs scenario 3 for seasonal storages with various capacities. For each runs stores investment cost, operational cost and emissions.

Points that need discussion:

    1. How to best implement half hour time steps and cost of storage charging
    2. On what gas price to decide on - choice has major effects on results

```python
import matplotlib.pyplot as plt
plt.style.use('bmh')

from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv())
import os
import sys

PROJECT_ROOT = os.environ.get('PROJECT_SRC')
sys.path.append(PROJECT_ROOT)

from easter_bush_energy.scenario_2 import run_scenario_2
from easter_bush_energy.scenario_3 import run_scenario_3

storage_e_noms = [200_000, 300_000, 400_000, 500_000]

# other fixed parameters
dT_lts = 40
dT_sts = 40

start = '2019-01-01'
end = '2020-01-01'

investment_costs = list()
operating_costs = list()
total_costs = list()
emissions = list()
emissions_costs = list()

plots = [
    'investment_cost [Pound]',
    'operating_cost [Pound]',
    'total_cost [Pound]',
    'emission [kg]',
    'cost per saved emission [Pound/kg]'
    ]

base_results = run_scenario_2(start=start, end=end)
base_emissions = base_results['gas_emission'] + base_results['elec_emission']
base_opcosts = base_results['operating_cost']

for e_nom in storage_e_noms:
    results = run_scenario_3(storage_e_nom=e_nom, start=start, end=end)

    investment_costs.append(results['total_investment'])
    operating_costs.append(results['operating_cost'])
    ems = results['gas_emission'] + results['elec_emission']
    emissions.append(ems)
    total_cost = investment_costs[-1] + operating_costs[-1]
    total_costs.append(total_cost)

    delta_em = base_emissions - ems
    delta_cost = base_opcosts - total_cost
    cost_per_saved = abs(delta_cost / delta_em)

    emissions_costs.append(cost_per_saved)


# {'gas_used': 393541, 'elec_used': 1019907, 'gas_emission': 72412, 'elec_emission': 24478, 
# 'operating_cost': 52500, 'heatpump2stes_investment': 929192.8804386624, 'total_investment': 929192.8804386624}

plt.show()
fig, axs = plt.subplots(1, 5, figsize=(25, 6))

axs[3].plot([min(storage_e_noms)-100, max(storage_e_noms)+100], 
                    [base_emissions, base_emissions],
                    c='r', alpha=0.4, label='Scenario 2')
axs[1].plot([min(storage_e_noms)-100, max(storage_e_noms)+100], 
                    [base_opcosts, base_opcosts],
                    c='r', alpha=0.4)
axs[2].plot([min(storage_e_noms)-100, max(storage_e_noms)+100], 
                    [base_opcosts, base_opcosts],
                    c='r', alpha=0.4)

data_list = [investment_costs, operating_costs, total_costs, emissions, emissions_costs]
for data, ax, ylabel in zip(data_list, axs, plots):
    
    ax.scatter(storage_e_noms, data, label='Scenario 3', linewidths=0.5, edgecolor='k', color='darkred', s=20) 
    ax.set_ylabel(ylabel)
    ax.set_xlabel('STES Storage Capacity')
    ax.set_xlim(min(storage_e_noms)-3000, max(storage_e_noms)+3000)

axs[0].legend()
plt.tight_layout()
plt.show()
```
